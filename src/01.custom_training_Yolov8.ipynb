```python
from ultralytics import YOLO
import cv2
import torch
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'
```

```python
# 모델 설정 및 학습 // 경로 주의
# 모델 설정
model = YOLO("yolov8s.pt") 

# 모델 학습 // roboflow에서 다운받은 yaml 파일 이름은 'data.yaml' >> 임의로 변경하였음
model.train(data='./motorcycle_rider3.yaml' , epochs=24, name='8ne100_v3') # name은 모델이 저장될 파일 이름, drop_out=0.0(default) >> 과적합 관리
```

```python
# 학습된 모델의 클래스 확인
model.names
```

```python
# 결과 확인
# model = YOLO('./runs/detect/8le100/weights/best.pt') # 학습된 가중치 불러오기

path_video = '../final/Video Data 2/Morning.mp4' # 영상 경로

cap = cv2.VideoCapture(path_video)

while cap.isOpened():
    success, frame = cap.read()

    if success:
        results = model(frame)

        annotated_frame = results[0].plot()

        cv2.imshow("result", annotated_frame)

        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()
```
